```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import insightface
from insightface.app import FaceAnalysis
```

```python
app = FaceAnalysis(name="buffalo_s",providers=['CPUExecutionProvider'])
app.prepare(ctx_id=0,det_size=(640,640))
```

```python
imag_2 = cv2.imread('input/Mohmmad.png')
imag_2 = cv2.cvtColor(imag_2,cv2.COLOR_BGR2RGB)
plt.imshow(imag_2)
plt.show()
```

```python
result_2 = app.get(imag_2)
embedding_1 = result_2[0]["embedding"].shape
```

```python
imag_3 = cv2.imread('input/Sajjad.jpeg')
imag_3 = cv2.cvtColor(imag_3,cv2.COLOR_BGR2RGB)
plt.imshow(imag_3)
plt.show()
```

```python
result_2 = app.get(imag_2)
embedding_2 = result_2[0]["embedding"].shape
```

```python
np.sqrt(np.sum((embedding_1 - embedding_2) **2))
```
